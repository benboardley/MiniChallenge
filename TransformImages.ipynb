```python
!pip install torchvision
from google.colab import drive
from torchvision import transforms
from PIL import Image
import torch
import os
from pathlib import Path
import glob
from tqdm.auto import tqdm


drive.mount('/content/drive')


celebrities_dir = '/content/drive/MyDrive/ECE 50024/cropped_large_outlier_excluded'
save_dir = '/content/drive/MyDrive/ECE 50024/cropped_large_tensors_outlier_excluded'


os.makedirs(save_dir, exist_ok=True)


def preprocess_and_save_images(image_paths, save_dir):
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5957162,  0.4629442,  0.40489063], std=[0.22739226, 0.20087235, 0.18877847])
    ])

    for image_path in tqdm(image_paths, desc='Processing Images'):
        img = Image.open(image_path).convert('RGB')
        img = transform(img)
        filename = Path(image_path).stem + '.pt'  # Get the file name without extension
        torch.save(img, os.path.join(save_dir, filename))


image_files = glob.glob(f"{celebrities_dir}/*.jpg")


preprocess_and_save_images(image_files, save_dir)

print("Preprocessing and saving completed.")
```
