```python
!pip install facenet-pytorch
!pip install torchvision
import os
from PIL import Image
import shutil
from tqdm import tqdm
from google.colab import drive
from facenet_pytorch import MTCNN

drive.mount('/content/drive')

mtcnn = MTCNN(keep_all=True)

base_dir = '/content/drive/MyDrive/ECE 50024/test'


output_dir_cropped = '/content/drive/MyDrive/ECE 50024/cropped_test'
output_dir_outliers_excluded = False#'/content/drive/MyDrive/ECE 50024/cropped_large_outlier_excluded'


if os.path.exists(output_dir_cropped):
    shutil.rmtree(output_dir_cropped)
os.makedirs(output_dir_cropped)



def preprocess_image_facenet(image_path, mtcnn, save_path, save_outlier_excluded_path=None):
    img = Image.open(image_path).convert('RGB')
    boxes, _ = mtcnn.detect(img)
    if boxes is not None and len(boxes) > 0:

        box = boxes[0].astype(int)
        cropped_img = img.crop((box[0], box[1], box[2], box[3]))
        cropped_img.save(save_path)
        if save_outlier_excluded_path:
            cropped_img.save(save_outlier_excluded_path)
        return True
    else:

        img.save(save_path)
        return False


for image_name in tqdm(os.listdir(base_dir), desc=f"Processing"):
        image_path = os.path.join(base_dir, image_name)

        save_path_cropped = os.path.join(output_dir_cropped, image_name)


        preprocess_image_facenet(image_path, mtcnn, save_path_cropped)#, save_path_outlier_excluded)

print("Processing complete.")
```
